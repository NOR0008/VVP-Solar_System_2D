Aktuální pozice planet na začátku, v tomto stavu pohyb ještě nezačal. 
Na prvním obrázku jsou všechny objekty solárního systému. Druhý obrázek přibližuje pět objektů solárního systému, které jsou na prvním obrázku "pohromadě".

```python
%matplotlib widget
SolSys = Solar_system(json_data('data2D/planets.json'))
SolSys.draw_place(7e12)
SolSys.draw_place(7e11)
```

Simulace: časový průběh planet za 1 rok, tj.365 dní. 86400 je počet sekunt za 1 den. To znamená: 60 sekund * 60 minut * 24 hodin.

```python
SolSys.progressing_simulate(86400*365, 86400)
SolSys.draw_way(7e12)
```

Simulace: místo planet v solárním systému a časový průběh trajektorií za pět let:

```python
SolSys.progressing_simulate(86400*365*5, 86400)
SolSys.draw_way(7e12)
SolSys.animation_planets(5e11)
```

Simulace: časový průběh trajektorií a místo planet za deset let:

```python
SolSys.progressing_simulate(86400*365*10, 86400)
SolSys.draw_way(7e12)
SolSys.animation_planets(5e12)
```
